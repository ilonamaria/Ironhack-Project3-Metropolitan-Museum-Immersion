```python
# FURTHER ANALYSIS

# 1. Use NLP to identify words clusters and compare to words groups I created.Reference:
# https://www.kaggle.com/kitakoj18/exploring-wine-descriptions-with-nlp-and-kmeans
# I tried a more automatic way to group words, applying Kmeans, but the result wasn't too acurate:

from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from nltk.tokenize import RegexpTokenizer
from nltk.stem.snowball import SnowballStemmer
%matplotlib inline
```

```python
punc = ['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}',"%"]
stop_words = text.ENGLISH_STOP_WORDS.union(punc)
desc = df_ajustado['title_lower'].values
vectorizer = TfidfVectorizer(stop_words = stop_words)
X = vectorizer.fit_transform(desc)
```

```python
word_features = vectorizer.get_feature_names()
word_features[1:10000]
```

```python
stemmer = SnowballStemmer('english')
tokenizer = RegexpTokenizer(r'[a-zA-Z\']+')

def tokenize(text):
    return [stemmer.stem(word) for word in tokenizer.tokenize(text.lower())]
```

```python
vectorizer2 = TfidfVectorizer(stop_words = stop_words, tokenizer = tokenize)
X2 = vectorizer2.fit_transform(desc)
word_features2 = vectorizer2.get_feature_names()
word_features2[:5000]
```

```python
vectorizer3 = TfidfVectorizer(stop_words = stop_words, tokenizer = tokenize, max_features = 1000)
X3 = vectorizer3.fit_transform(desc)
words = vectorizer3.get_feature_names()
```

```python
kmeans = KMeans(n_clusters = 20, n_init = 5, n_jobs = -1)
kmeans.fit(X3)
```

```python
common_words = kmeans.cluster_centers_.argsort()[:,-1:-22:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```

```python
df_ajustado['cluster'] = kmeans.labels_
```

```python
#Non accuracy factor was visualized in the heat map below:

clusters = df_ajustado.groupby(['cluster', 'department']).size()
fig2, ax2 = plt.subplots(figsize = (30, 15))
sns.heatmap(clusters.unstack(level = 'department'), ax = ax2, cmap = 'Reds')

ax2.set_xlabel('Department', fontdict = {'weight': 'bold', 'size': 16})
ax2.set_ylabel('Cluster', fontdict = {'weight': 'bold', 'size': 16})
for label in ax2.get_xticklabels():
    label.set_size(14)
    label.set_weight("bold")
for label in ax2.get_yticklabels():
    label.set_size(14)
    label.set_weight("bold")
```

```python
import re
  
def remove(list):
    pattern = '[0-9]'
    list = [re.sub(pattern, '', i) for i in list]
    return list
  
# Driver code 
  
list = all_titles
print(remove(list))
```

```python
!pip3 install stop_words
from stop_words import get_stop_words
```

```python
comment_words = ''
stop_words = ['the', "the'",'in', 'mr','on','c', 'of', 'a',"a'", "de'",'A', 'B',"'",'and', 'with', "with'","from'",'from', 'an', 'at', 'to', 'by', 'for', 'e', 'S']

# iterate through the csv file
for word in remove(list):
      
    # typecaste each val to string
    word = str(word)
  
    # split the value
    tokens = word.split()
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1200, height = 800, 
                background_color ='white',
                stopwords = stop_words,
                min_font_size = 10).generate(comment_words)
  
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
